```python
# 1. Making time array till 50 with intervals of 0.01
t = np.arange(0,50,0.01)
# 2. Calculating 3 functions in this time interval
f = np.sin(np.pi * t)
g = np.cos(np.pi*0.1*t)
i = f*g

# NOTE :: Array computations are simplified with Numpy package.
# If you want, try these computations only using Python Lists to see the difficulty of performing computation
# without numpy
```

```python
# 1. Random number arrays : x_ran, y_ran
N = 10000
xran = np.random.uniform(0,1,N)
yran = np.random.uniform(0,1,N)
xvalue = np.linspace(0,1,N)
# 4. Calculating the sinfunc
sinfunc = np.sin(np.pi*xvalue)
```

```python
# 2. and 5. Plotting the Scatter plot of x_ran, y_ran and also the sinfunc
plt.scatter(xran,yran,marker='x')
plt.plot(xvalue,sinfunc,c='r',lw=2)
plt.show()
#check the parameters passed above, 
# marker :: type of marker to use, try 'o', '*', or look up other kinds of marker available
# c :: this is the color of line, you can also use `color` instead of `c`, try, 'b', 'k', or other colors
# lw :: this is the width of the line, you can also use `linewidth` instead of `lw`. try lw = 0, 1,2,3,4...
# there are many more paramters available to make plots pretty, you can check them up, 
# try to look up for plt.plot parameters
```

```python
# 6. Counting the number of points inside the sin curve
# First : creating a list of tuples of x,y coordinate. this will look like :  [(x1,y1),(x2,y2),.....(xN,yN)]
coordinate = zip(xran,yran) 
# counter to count the number of points inside the sine
counter = 0
for x,y in coordinate:
    if y<np.sin(np.pi*x):# Checking if given point is inside the sine curve
        counter += 1 # if yes, counting that point
print("Number of points inside the sine curve : %d"%counter)
```

```python
# Now calculating the ratio of points inside to total number of points
ratio = float(counter)/float(N)
print("Our Ratio of points inside the curve is %f"%ratio)
```

# Note on Task 2 : 

 What we did was motecarlo integration of $sin(\pi x)$ function in the domain $x \in [0,1]$   
 Verify the following :    
 
 $$\int_0^1 sin(\pi x) dx =  \frac{2}{\pi} = 0.6366197$$    
 
 Now, look back at what we did for Task 2.  
  1. We took N random points in the square box (x going from 0 to 1, and y going from 0 to 1)
  2. We then counted the number points from this N random points that fall under our sine curve
  3. We calculated the ratio of points that fall under curve to total points.
In this way, we calculated what fraction of square box is taken up by our sine curve. We sampled the square box by taking N random points in $x\in[0,1]$, $y\in[0,1]$) and counted the fraction taken up by sine curve by counting the points that fall under the curve.  

Our formula for calculating the area under our sine curve was actually, 

$$area_under_curve = \frac{Number\_of\_points\_under\_curve}{N}*Area\_of\_our\_total\_sampled\_space$$   

As we sampled a unit square (square of side 1, as our x and y went from 0 to 1), we could simply our calculation by,
$$Area\_of\_our\_total\_sampled\_space = 1$$  

But say, you calcuate sin curve by sample points that goes is in $x \in [0,2]$, $y \in [0,2]$, then you have sampled square that has side of length $2$. So you have to calculate area under curve by multiply by $Area\_of\_our\_total\_sampled\_space = 2$ with the calculated ratio of points under curve.  

For more information checkout : https://en.wikipedia.org/wiki/Monte_Carlo_integration
